```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()

import numpy as np
```

```python
from bokeh.plotting import figure, output_file, show
```

```python
# Q1) Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x

output_file("sin_wave.html")

X = np.arange(0, 10)
Y = np.sin(X)

p = figure(title="Sine Wave", x_axis_label="X", y_axis_label="Y")
p.line(X, Y, legend_label="Sine Wave", line_width=2)

show(p)
```

```python
# Q2) Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.

n = 100
X = np.random.rand(n)
Y = np.random.rand(n)

sizes = np.random.randint(1, 50, n)
colors = np.random.choice(['red','blue','brown','yellow','green'], size=n)

output_file("scatter_plot.html")

p = figure(title="Scatter Plot", x_axis_label="X", y_axis_label="Y")
p.scatter(X, Y, size=sizes, color=colors, alpha=0.5)
show(p)
```

```python
# Q3)  Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

output_file("bar_chart.html")

p = figure(x_range=fruits, title="Fruit Counts")
p.vbar(x=fruits, top=counts, width=0.5)

show(p)
```

```python
# Q4) Create a Bokeh histogram to visualize the distribution of the given data

data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

output_file("histogram.html")

p = figure(title="Histogram", x_axis_label="Value", y_axis_label="Frequency")
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="navy", line_color="white", alpha=0.5)

show(p)
```

```python
# Q5) Create a Bokeh heatmap using the provided dataset.

data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)


output_file("heatmap.html")

p = figure(title="Heatmap", x_axis_label="X", y_axis_label="Y")
p.image(image=[data_heatmap])

show(p)
```
